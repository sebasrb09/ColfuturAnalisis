```python
def find_city(uni):
    uni = str(uni).replace("University","")
    uni = str(uni).replace("univerisity","")
    uni = str(uni).replace("U.","")
    uni = str(uni).replace("School","")
    uni = str(uni).replace("Science","")
    uni = str(uni).replace("of","")
    doc = nlp(uni)
    res = None
    print(doc)
    for ent in doc.ents:
        if ent.label_ == 'GPE':
            res = ent.text
    return res
```

```python
def find_city3(uni):
    uni = str(uni).replace("University","")
    uni = str(uni).replace("Université","")
    uni = str(uni).replace("univerisity","")
    uni = str(uni).replace("U.","")
    uni = str(uni).replace("School","")
    uni = str(uni).replace("Science","")
    uni = str(uni).replace("of","")
    text_arr = uni.split()
    res = None
    for i in text_arr:
        doc = nlp(i)
        for ent in doc.ents:
            if (res is None):
                res = ent.text
    return res
```

```python
def find_city2(uni):
    text_arr = uni.split()
    res = None
    for i in text_arr:
        doc = nlp(i)
        for ent in doc.ents:
            if ent.label_ == 'GPE' and (res is None):
                res = ent.text
    return res
```
